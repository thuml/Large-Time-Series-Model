```python
import torch
seq_len = 768
batch_size = 768
n_features = 12
x_mark_enc = torch.arange(seq_len, dtype=torch.float32).repeat(batch_size, n_features, 1).permute(0, 2, 1)
print(x_mark_enc.shape)

pred_len = 96
x_mark_dec = torch.arange(seq_len, seq_len + pred_len, dtype=torch.float32).repeat(batch_size, n_features, 1).permute(0, 2, 1)
print(x_mark_dec.shape)
```
